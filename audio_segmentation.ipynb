```python
from pydub import AudioSegment

PROJECT_HOME = './'
TRAINING_VALIDATION_TEST_DATASET_DIR = PROJECT_HOME + 'dataset/9. training_validation_test (w.o)/'
RESULT_DIR = PROJECT_HOME + 'dataset/10. segmentation (w.o)/'

sec = 1000
b_with_video_link = False
```

```python
# AutelEVO
if b_with_video_link == True:
    file_num = 0

    for count in range (1, 4):
        audio_file = AudioSegment.from_wav(TRAINING_VALIDATION_TEST_DATASET_DIR + 'AutelEVO_Test_File' + str(count) + '.wav')
        iter_num = int(len(audio_file)/1000) # round(len(audio_file)/1000)

        for i in range (0, iter_num):
            file = audio_file[i*sec: (i+1)*sec]
            file.export(RESULT_DIR + 'AutelEVO_' + f"{file_num:04d}" + '.wav', format="wav")
            file_num = file_num + 1       
        
    for count in range (1, 6):
        audio_file = AudioSegment.from_wav(TRAINING_VALIDATION_TEST_DATASET_DIR + 'AutelEVO_Training_File' + str(count) + '.wav')
        iter_num = int(len(audio_file)/1000) # round(len(audio_file)/1000)

        for i in range (0, iter_num):
            file = audio_file[i*sec: (i+1)*sec]
            file.export(RESULT_DIR + 'AutelEVO_' + f"{file_num:04d}" + '.wav', format="wav")
            file_num = file_num + 1       

    for count in range (1, 4):
        audio_file = AudioSegment.from_wav(TRAINING_VALIDATION_TEST_DATASET_DIR + 'AutelEVO_Validation_File' + str(count) + '.wav')
        iter_num = int(len(audio_file)/1000) # round(len(audio_file)/1000)

        for i in range (0, iter_num):
            file = audio_file[i*sec: (i+1)*sec]
            file.export(RESULT_DIR + 'AutelEVO_' + f"{file_num:04d}" + '.wav', format="wav")
            file_num = file_num + 1       
```
