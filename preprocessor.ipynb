```python
def get_integrated_signal(target_dir=PROJECT_HOME, type='') -> np.ndarray:
    arr_drone_paths = librosa.util.find_files(target_dir, ext='wav')
    arr_drone_paths = np.asarray(arr_drone_paths)

    arr_drone_signal = np.array([])
    dict_drone_signal = {}    
    
    next_path_index = 0
    number_of_file = 0

    for path in arr_drone_paths:        
        signal, sr = librosa.load(path, sr=SAMPLE_RATE)        
        arr_drone_signal = np.append(arr_drone_signal, signal)   

        cur_drone_name = path.split('\\')[-1].split('_')[0]

        print(type + " / " + cur_drone_name + "'s signal : ", signal.shape)
        print(type + " / " + cur_drone_name + "'s integrated signal : ", arr_drone_signal.shape)
               
        next_path_index += 1
        number_of_file += 1

        if next_path_index < len(arr_drone_paths):
            next_drone_name = arr_drone_paths[next_path_index].split('\\')[-1].split('_')[0]                                
        else:
            next_drone_name = ' '
        
        if cur_drone_name != next_drone_name:
            dict_drone_signal[cur_drone_name] = arr_drone_signal
            print("Number of " + cur_drone_name + "'s files : ", number_of_file)
            print("Adding " + cur_drone_name + "'s integrated signal : ", arr_drone_signal.shape)
            arr_drone_signal = np.array([])
            number_of_file = 0
        
    return dict_drone_signal


# SEGMENTATION_DATASET_DIR = PROJECT_HOME + 'dataset/6. segmentation/'
# dict_integrated_train_validation_test_signal = get_integrated_signal(SEGMENTATION_DATASET_DIR, type='TRAINING+VALIDATION+TEST')
```

```python
def get_all_files(target_dir=PROJECT_HOME) -> list:   
    list_files = []

    # root : 하부 디렉토리들, dirs : root 아래 디렉토리 리스트, files : root 아래 파일 목록
    for (root, dirs, files) in os.walk(target_dir):        
        if root == target_dir or root == target_dir + '/':
            for file in files:
                ext = os.path.splitext(file)[-1]
                if ext == '.WAV' or ext == '.wav':
                    list_files.append(root + '/' + file)

    return list_files
```

```python
def get_target_names(target_dir=PROJECT_HOME) -> list:
    # list_train_drone_paths = get_all_files(TRAINING_DATASET_DIR)
    # list_validation_drone_paths = get_all_files(VALIDATION_DATASET_DIR)
    list_test_drone_paths = get_all_files(target_dir)

    list_drone_ids = []

    for drone_path in list_test_drone_paths:
        drone_id = drone_path.split('/')[-1].split('_')[0]

        if drone_id not in list_drone_ids:
            list_drone_ids.append(drone_id)

    # for item in range(0, len(list_drone)):
    #     print(list_drone[item])
        
    return list_drone_ids
```
